```python
import Model
import DataSet
import tensorflow as tf
import os
```

```python
def test_accuracy():
    corr = 0
    for images, labels in DataSet.iter_batches(500, DataSet.TEST_DATASET):
        corr += sum(model.correct.eval(feed_dict = {
            model.input_var: images,
            model.corr_labels: labels,
            model.keep_prob: 1.0,
            model.use_precomputed_means: True,
        }))
    return float(corr)/len(DataSet.TEST_DATASET.images)
```

```python
model = Model.Model()
summary = tf.merge_all_summaries()
session = tf.Session()
session.run(tf.initialize_all_variables())

writer = tf.train.SummaryWriter('summary-log', session.graph_def)

with session.as_default():
    counter = 0
    for epoch in range(10):
        print(epoch)
        for images, labels in DataSet.iter_batches(30):
            counter += 1
            if counter % 100 == 0:
                acc, summ = session.run([model.accuracy, summary], feed_dict = {
                    model.input_var: images,
                    model.corr_labels: labels,
                    model.keep_prob: 1.0,
                    model.use_precomputed_means: True,
                })

                writer.add_summary(summ, counter)
                print("iteration {}, training accuracy {}".format(counter, acc))

            model.train.run(feed_dict = {
                model.input_var: images,
                model.corr_labels: labels,
                model.keep_prob: 0.5,
                model.use_precomputed_means: False,
            })

        print("After epoch {}: test accuracy: {}".format(epoch, test_accuracy()))
        writer.flush()
```
